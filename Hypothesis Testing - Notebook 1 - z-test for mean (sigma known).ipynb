#                       Steps in conducting Hypothesis Testing:

### Example 1: Z-test (sigma of the population is known)

#### Z statistic = ((Xbar - mu)/(sigma/sqrt(n))
##### All of the parameters follow the usual nomenclature.

##### Q1. The quality-control manager at a compact fluorscent light bulb (CFL) factor needs to determine whether the mean life of a large shipment of CFLs is equal to 7500 hours. The quality-control manager has to check if the randomly sampled CFL bulbs are in alignment with the promises made by the company to their customers.  The claim by the company is that the standard deviation is 1000 hours. A random sample of 64 CFLSs indicates a sample mean life of 7250 hours.


###### a) At 0.05 level of significance, is there evidence that the mean life is different from 7500 hours?

Let us formulate the hypothesis.

H0 (null hypothesis):mu (population mean) = 7500

H1 (alternate hypothesis): mu (population mean) not equal to 7500

[this is an example of a two-tailed test]

```python
import numpy as np
import scipy
```

```python
z = ((7250-7500)/(1000/np.sqrt(64)))
z
```

At alpha = 0.025, the value of the z-statistic from the table is 1.96.
Here, we are taking the value of alpha as 0.025 since this a two-tailed test. So, the boundary of values can be said to lie between (-1.96,+1.96) with a confidence level of 95%.

Thus, the value of the z-statistic calculated from the data given does not fall in the range of (-1.96,+1.96).

Hence, we are rejecting H0.
mu (population mean) is not equal to 7500.

###### b) Compute the p-value and interpret its meaning.

Now, let us calculate the p value from the corresponding z-statistic and compare it with the level of significance (alpha = 0.05.

```python
p_value = scipy.stats.norm.sf(abs(z))*2 #mulitplying my 2 since this is a two-tailed test
p_value
```

We see that the p_value<z.
Thus, it is confirmed we are failing to accept the the null hypothesis since we have less than 5% probability that the null hypothesis will be correct.

###### c) Construct  a 95% confidence interval estimate of the population mean life of CFLS.

The mathematical formula for computing a 95% confidence interval is ((X(bar)-1.96/(sigma(sqrt(n))),(X(bar)+1.96(sigma/(sqrt(n))))

For the Left Hand side Limit of the confidence interval we have:

```python
L = 7250-((1.96)*((1000)/np.sqrt(64)))
L
```

For the Right Hand side Limit of the confidence interval we have:

```python
R = 7250+((1.96)*((1000)/np.sqrt(64)))
R
```

Thus, we can say we with 95% certainty that the population mean will lie in the interval of (7005,7495).
