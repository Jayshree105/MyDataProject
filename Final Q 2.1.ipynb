```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
#Variable with outliers WorkingHoursWife', 'EducationWife', 'WifeHourEarnings', 'WifeWage', 'WorkingHoursHusband', 'EducationHusband', 'HusbandWage', 'WifeExperience', 'FamilyIncome'

lrWH_wife,urWH_wife=remove_outlier(income['WorkingHoursWife'])
income['WorkingHoursWife']=np.where(income['WorkingHoursWife']>urWH_wife,urWH_wife,income['WorkingHoursWife'])

lrwife_edu,urwife_edu=remove_outlier(income['EducationWife'])
income['EducationWife']=np.where(income['EducationWife']>urwife_edu,urwife_edu,income['EducationWife'])
income['EducationWife']=np.where(income['EducationWife']<lrwife_edu,lrwife_edu,income['EducationWife'])

lrwife_ern,urwife_ern=remove_outlier(income['WifeHourEarnings'])
income['WifeHourEarnings']=np.where(income['WifeHourEarnings']>urwife_ern,urwife_ern,income['WifeHourEarnings'])

lrwife_wage,urwife_wage=remove_outlier(income['WifeWage'])
income['WifeWage']=np.where(income['WifeWage']>urwife_wage,urwife_wage,income['WifeWage'])

lrWH_hsb,urWH_hsb=remove_outlier(income['WorkingHoursHusband'])
income['WorkingHoursHusband']=np.where(income['WorkingHoursHusband']<lrWH_hsb,lrWH_hsb,income['WorkingHoursHusband'])
income['WorkingHoursHusband']=np.where(income['WorkingHoursHusband']>urWH_hsb,urWH_hsb,income['WorkingHoursHusband'])

lrhsb_edu,urhsb_edu=remove_outlier(income['EducationHusband'])
income['EducationHusband']=np.where(income['EducationHusband']<lrhsb_edu,lrhsb_edu,income['EducationHusband'])

lrhsb_wage,urhsb_wage=remove_outlier(income['HusbandWage'])
income['HusbandWage']=np.where(income['HusbandWage']>urhsb_wage,urhsb_wage,income['HusbandWage'])

lrwife_exp,urwife_exp=remove_outlier(income['WifeExperience'])
income['WifeExperience']=np.where(income['WifeExperience']>urwife_exp,urwife_exp,income['WifeExperience'])

lrfam_inc,urfam_inc=remove_outlier(income['FamilyIncome'])
income['FamilyIncome']=np.where(income['FamilyIncome']>urfam_inc,urfam_inc,income['FamilyIncome'])
```

**SCALING OF DATA**
Often the variables of the data set are of different scales i.e. one variable is in millions and other in only 100. For e.g. in our data set Income is having values in thousands and age in just two digits. Since the data in these varaibles are of different scales, it is tough to compare these variables.

Feature scaling (also known as data normalization) is the method used to standardize the range of features of data. Since, the range of values of data may vary widely, it becomes a necessary step in data preprocessing while using machine learning algorithms.

In this method, we convert variables with different scales of measurements into a single scale.

StandardScaler normalizes the data using the formula (x-mean)/standard deviation.

We will be doing this only for the numerical variables.

```python
from sklearn.preprocessing import StandardScaler
std_scale = StandardScaler()
income_std=std_scale.fit_transform(income)
```
